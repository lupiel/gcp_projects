```python
import seaborn as sns
```

```python
df = sns.load_dataset("diamonds")
```

```python
df
```

```python
for group_name, df_group in df.groupby(["cut", "color", "clarity"]):
    df_group.reset_index(inplace=True, drop=True)
    print(group_name)
    print(df_group)
    break
```
